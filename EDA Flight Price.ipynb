```python
def clean_and_convert(val):
    try:
        val = ''.join(filter(str.isdigit, str(val)))
        return int(val) if val != '' else np.nan
    except:
        return np.nan

df['Duration_hours'] = df['Duration_hours'].apply(clean_and_convert)

df['Duration_hours'] = df['Duration_hours'].fillna(0).astype(int)
```
